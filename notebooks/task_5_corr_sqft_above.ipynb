```python
# Task 5: Correlation of sqft_above with Price
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned dataset
file_path = 'C:/Users/oscar/OneDrive/Escritorio/Python_Projects/House_Sales_Project/data/kc_house_data_cleaned.csv'
df = pd.read_csv(file_path)

# Create a regression plot to see the correlation
sns.regplot(x='sqft_above', y='price', data=df, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('Correlation between sqft_above and Price')
plt.xlabel('Square Feet Above')
plt.ylabel('Price')
plt.show()
```
